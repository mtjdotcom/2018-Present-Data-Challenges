```python
df['Exp_ordinal'] = ''

for experience in df['Experience']:
    if experience == 0:
        df['Exp_ordinal'] = "Rookie"
    if experience > 0 and experience <= 3:
        df['Exp_ordinal'] = "Little Experience"
    if experience > 3 and experience <= 5:
        df['Exp_ordinal'] = "Experienced"
    if experience > 5 and experience <= 10:
        df['Exp_ordinal'] = "Very experienced"
    if experience > 10:
        df['Exp_ordinal'] = "Veteran"
```
