```python
import pandas as pd
```

```python
artworks = pd.read_csv("moma_artworks_2.csv")
artworks.head(20)
```

```python
# Let's use a Pandas feature called value_counts(), which returns a list of all the values in that column in addition 
# to their frequency of appearance, similar to a histogram. We want to make sure we have a good sense of all the 
# different formats the values in the Date column take on.

artworks["Date"].value_counts()
```

```python
def clean_split_dates(row):
    # Initial date contains the current value for the Date column for that row
    initial_date = str(row['Date'])
    # Same as the current value for the Date column for now. Update if a "-" is found!
    final_date = initial_date
    
    # .split('-') : Looks for a dash, if it finds it, the split the string on the dash,
    # and create a list with the values surrounding the dash, called `split_date`
    # E.g. : "1976-97" becomes ["1976", "97"], and we can just grab the first value
    split_date = initial_date.split('-') 

    # If no "-" found, split_date will just contain one item, the initial_date
    # If a "-" is found, split_date will contain a list of at least 2 elements
    if len(split_date) > 1:
        # In programming, the 0th element is the 1st element
        final_date = split_date[0]
    
    # If no dash was found, final_date is the same as initial_date
    return final_date

# Assign the results of "clean_split_dates" to the 'Date' column. Since we want Pandas to go
# row by row, we set "axis=1". If we wanted to go by columns, we would use "axis=0".
artworks['Date'] = artworks.apply(lambda row: clean_split_dates(row), axis=1)
artworks['Date'].value_counts()
```

```python
def clean_c_dates(row):
    initial_date = row['Date']
    # Use .lstrip() to strip from the left side - "c. " - if it exists
    final_date = initial_date.lstrip("c. ")
    return final_date

artworks["Date"] = artworks.apply(lambda row: clean_c_dates(row), axis=1)
artworks["Date"].value_counts()
```

```python
def clean_nd_dates(row):
    initial_date = row['Date']
    final_date = initial_date
    
    # Use "==" to check for equality
    if initial_date == "n.d.":
        final_date = "Unknown"

    # If not equal to "n.d.", keep original value
    return final_date

# Assign the results of "clean_nd_dates" as applied over every row, to the 'Date' column
artworks['Date'] = artworks.apply(lambda row: clean_nd_dates(row), axis=1)
artworks['Date'].value_counts()
```

```python
artworks["ArtistBio"].value_counts()
```

```python
def clean_parentheses(row):
    initial_bio = str(row['ArtistBio'])
    # Default value for final_bio is initial_bio
    final_bio = initial_bio
    
    right_strip = initial_bio.lstrip("(")
    left_strip = right_strip.rstrip(")")
    
    final_bio = left_strip
    
    return final_bio

# Remove `#` from the line below when ready to test!
artworks['ArtistBio'] = artworks.apply(lambda row: clean_parentheses(row), axis=1)
```
