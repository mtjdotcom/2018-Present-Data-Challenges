## Market Comparison of Google, Facebook, Microsoft

This analysis compares the volatility of Goog, FB, and MSFT

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
msft_stock = pd.read_csv("data/msft.csv")
goog_stock = pd.read_csv("data/goog.csv")
apple_stock = pd.read_csv("data/aapl.csv")
msft_stock.head(5)
```

```python
msft_stock["Date"] = pd.to_datetime(msft_stock["Date"])
goog_stock["Date"] = pd.to_datetime(goog_stock["Date"])
apple_stock["Date"] = pd.to_datetime(apple_stock["Date"])
```

```python
msft_date_type = msft_stock.dtypes["Date"]
goog_date_type = goog_stock.dtypes["Date"]
apple_date_type = apple_stock.dtypes["Date"]
str(msft_date_type)
```

```python
# plt.plot() tells Python to create a new plot and what the X (first argument) and Y (second argument) axes are
plt.plot(msft_stock["Date"], msft_stock["Close"])
# plt.title() details how we want the current plot to be titled
plt.title("Microsoft Daily Stock Price: Ending October 17, 2015")
plt.show()
```

```python
plt.plot(goog_stock["Date"], goog_stock["Close"])
plt.title("Google Daily Stock Price: Ending October 17. 2015")
plt.show()
```

```python
plt.plot(apple_stock["Date"], apple_stock["Close"])
plt.title("Apple Daily Stock Price: Ending October 17, 2015")
plt.show()
```

```python
plt.plot(msft_stock["Date"], msft_stock["Close"])
plt.plot(goog_stock["Date"], goog_stock["Close"])
plt.plot(apple_stock["Date"], apple_stock["Close"])
plt.show()
```

```python
plt.plot(msft_stock["Date"], msft_stock["Volume"], color="Blue")
plt.title("Microsoft Trading Volume")
plt.show()
```

```python
plt.plot(apple_stock["Date"], apple_stock["Volume"], color="Red")
plt.title("Apple Trading Volume")
plt.show()
```

```python
plt.plot(goog_stock["Date"], goog_stock["Volume"], color="Green")
plt.title("Google Trading Volume")
plt.show()
```

```python
# Filter down to specfic periods
# Filter dates to be greater than Jan 1, 1999 but less than Jan 1, 2002
msft_bubble = msft_stock[(msft_stock["Date"] > "1999-01-01") & (msft_stock["Date"] < "2002-01-01")]
plt.plot(msft_bubble["Date"], msft_bubble["Volume"])
# .gcf() - returns the current figure (or plot)
fig_msft = plt.gcf()
# autofmt_xdate():  auto-format X-axis to look nice
plt.gcf().autofmt_xdate()
plt.title("Microsoft Dot Com Volume")
plt.show()
```

```python
# Filter down to specfic periods
# Filter dates to be greater than Jan 1, 1999 but less than Jan 1, 2002
apple_bubble = apple_stock[(apple_stock["Date"] > "1999-01-01") & (apple_stock["Date"] < "2002-01-01")]
plt.plot(apple_bubble["Date"], apple_bubble["Volume"])
fig_apple = plt.gcf()
plt.gcf().autofmt_xdate()
plt.title("Apple Dot Com Volume")
plt.show()
```

```python
# add new column that is the difference between high and low
goog_stock["Price-Travel"] = goog_stock["High"] - goog_stock["Low"]
goog_stock.head(5)
```

```python
plt.plot(goog_stock["Date"], goog_stock["Price-Travel"])
plt.title("Google Volitility")
plt.show()
```

```python
# 
goog_volitility = goog_stock[(goog_stock["Date"] > "2014-01-01") & (goog_stock["Date"] < "2015-10-17")]
plt.plot(goog_volitility["Date"], goog_volitility["Price-Travel"])
fig_goog = plt.gcf()
plt.gcf().autofmt_xdate()
plt.title("Google 2015 Volitility")
plt.show()
```
